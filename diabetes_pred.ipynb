```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
print(data.shape)
```

```python
# check null values
data.isnull().sum()
```

```python
data['Outcome'].value_counts()
```

```python
x = data.drop('Outcome', axis=1)
y = data['Outcome']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# model architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
m1 = models.Sequential()

m1.add(layers.Dense(24, activation='relu', input_dim=8))
m1.add(layers.Dense(16, activation='relu'))
m1.add(layers.Dense(8, activation='relu'))
m1.add(layers.Dense(1, activation='sigmoid'))

m1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
m1.fit(x_train, y_train, epochs=100, validation_data=(x_test, y_test))
```

```python
ypred = m1.predict(x_test)
print(ypred)
```

```python
#sorting ones and zeroes
ypred_m1 = [1 if i>0.5 else 0 for i in ypred]
print(ypred_m1)
```

```python
print(y_test.values)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, ypred_m1)
print(cm)
```
